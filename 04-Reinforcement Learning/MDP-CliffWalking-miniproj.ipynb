```python
import numpy as np
import gym
```

```python
env = gym.make('CliffWalking-v0')
env.reset()
env.render()
```

**Inisialisasi Tabel Q dan nilai discount factor**

```python
q_table = np.zeros((env.observation_space.n, env.action_space.n))
gamma = 0.9
```

```python
epsilon = 0.1
num_episodes = 500
max_steps_per_episode = 100
```

```python
def choose_action(state):
    if np.random.uniform(0, 1) < epsilon:
        return env.action_space.sample()  
    else:
        return np.argmax(q_table[state, :])
```

```python
for episode in range(num_episodes):
    state = env.reset()
    total_reward = 0
    done = False

    # Loop untuk menjalankan langkah-langkah di dalam episode
    for step in range(max_steps_per_episode):
        action = choose_action(state)
        new_state, reward, done, info = env.step(action)
        total_reward += reward

        # Update tabel Q berdasarkan persamaan Bellman
        q_table[state, action] = q_table[state, action] + gamma * (reward + np.max(q_table[new_state, :]) - q_table[state, action])

        state = new_state

        if done:
            np.save('q_table.npy', q_table)
            break

    print("Episode:", episode, "Total Reward:", total_reward)
```

```python
state = env.reset()
done = False
steps = []
while not done:
    action = np.argmax(q_table[state, :])
    new_state, _, done, _ = env.step(action)
    steps.append(new_state)
    state = new_state

env.render()
print("Jalur terpendek yang ditempuh:", steps)
```

```python
env.reset()
env.close()
```
